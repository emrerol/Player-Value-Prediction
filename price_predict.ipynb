```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
player_data = pd.read_csv('./variables.csv')
player_data.describe
```

```python
# Oyuncunun tahmini bonservis değer aralığını, minimum ve maximum olarak iki farklı sütuna ayırarak minimum bonservis bedeli, maksimum bonservis bedeli şeklinde ayırıyoruz

# Splitting the "Transfer value" column into two separate columns
player_data[['Min Transfer Value', 'Max Transfer Value']] = player_data['Transfer Value'].str.split('-', expand=True)

def convert_to_numeric(value):
    if value is not None:
        value = value.strip()  # Remove leading and trailing whitespaces
        if value.endswith('K'):
            return float(value.replace('€', '').replace('K', '')) * 1000
        elif value.endswith('M'):
            return float(value.replace('€', '').replace('M', '')) * 1000000
        else:
            return float(value.replace('€', ''))
    else:
        return None  # Return None for missing values

# Applying the conversion function to the columns
player_data['Min Transfer Value'] = player_data['Min Transfer Value'].apply(convert_to_numeric)
player_data['Max Transfer Value'] = player_data['Max Transfer Value'].apply(convert_to_numeric)

# Displaying the DataFrame
player_data.drop(columns=['Transfer Value'],inplace=True)
```

```python
player_data
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming df is your DataFrame and 'column_name' is the name of the column containing strings
column_name = 'Based'

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit LabelEncoder to the unique values in the column and transform the column
encoded_column = label_encoder.fit_transform(player_data[column_name])

# Print the mapping of original strings to encoded numbers
print("Mapping of original strings to encoded numbers:")
for label, string in enumerate(label_encoder.classes_):
    print(f"{string}: {label}")

# Replace the original column with the encoded column
player_data[column_name + '_encoded'] = encoded_column
player_data.drop(columns='Based', inplace=True)
```

```python
player_data
```

```python
columns_to_delete = ['Name', 'Height', 'Weight', 'Unnamed: 0']
player_data.drop(columns=columns_to_delete, inplace=True)

# NaN değerlerini ortalama ile doldurma
player_data = player_data.fillna(player_data.mean())

player_data.head()
```

```python
def assign_position(row):
    # Kaleci için kriterler
    if (row['Saves'] > 10 or row['Saves/xSv%'] > 0.1 or row['Sv %'] > 0.7 or 
        row['Svh'] > 10 or row['Svt'] > 20 or row['Pens Saved Ratio'] > 0.05):
        return 'Goalkeeper'
    # Defans için kriterler
    elif (row['Tck R'] > 0.5 or row['Int/90'] > 1.0 or row['Clear'] > 30 or 
          row['Hdr %'] > 30 or row['Fls'] > 10 or row['Aer A/90'] > 2):
        return 'Defender'
    # Orta Saha için kriterler
    elif (row['Pas %'] > 70 or row['Drb/90'] > 1.5 or row['Asts/90'] > 0.2 or 
          row['K Ps/90'] > 1.5 or row['Cr C/A'] > 0.3 or row['Tck R'] > 0.4):
        return 'Midfielder'
    # Forvet için kriterler
    elif (row['Gls'] > 10 or row['Gls/90'] > 0.3 or row['Shot %'] > 0.2 or 
          row['xG'] > 5 or row['Ch C/90'] > 1 or row['Drb/90'] > 3):
        return 'Forward'
    else:
        return 'Midfielder'  # Default to Midfielder if no other criteria are met

# Yeni mevki sütunu ekleme
player_data['Position'] = player_data.apply(assign_position, axis=1)

player_data
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoder oluşturma
le = LabelEncoder()

# Mevkileri sayısal değerlere dönüştürme
player_data['Position_encoded'] = le.fit_transform(player_data['Position'])

# Mevkilerin sayısal karşılıklarını yazdırma
print(dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
# Veriyi mevkilere göre ayırma
defender_data = player_data[player_data['Position_encoded'] == 0]
forward_data = player_data[player_data['Position_encoded'] == 1]
goalkeeper_data = player_data[player_data['Position_encoded'] == 2]
midfielder_data = player_data[player_data['Position_encoded'] == 3]

X_def = defender_data.drop(columns=['Value', 'Position_encoded'])
y_def = defender_data['Value']

X_fwd = forward_data.drop(columns=['Value', 'Position_encoded'])
y_fwd = forward_data['Value']

X_gk = goalkeeper_data.drop(columns=['Value', 'Position_encoded'])
y_gk = goalkeeper_data['Value']

X_mid = midfielder_data.drop(columns=['Value', 'Position_encoded'])
y_mid = midfielder_data['Value']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

def get_feature_importance(X, y):
    model = RandomForestRegressor(random_state=42)
    model.fit(X, y)
    importances = model.feature_importances_
    feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
    return feature_importance.sort_values(by='Importance', ascending=False)

# Her mevki için özellik önemlerini hesaplama
feature_importance_def = get_feature_importance(X_def, y_def)
feature_importance_fwd = get_feature_importance(X_fwd, y_fwd)
feature_importance_gk = get_feature_importance(X_gk, y_gk)
feature_importance_mid = get_feature_importance(X_mid, y_mid)

print("Defender Feature Importances:\n", feature_importance_def.head(10))
print("Forward Feature Importances:\n", feature_importance_fwd.head(10))
print("Goalkeeper Feature Importances:\n", feature_importance_gk.head(10))
print("Midfielder Feature Importances:\n", feature_importance_mid.head(10))
```

```python
def select_top_features(feature_importance, top_n=10):
    return feature_importance.head(top_n)['Feature'].values

top_features_def = select_top_features(feature_importance_def, top_n=10)
top_features_fwd = select_top_features(feature_importance_fwd, top_n=10)
top_features_gk = select_top_features(feature_importance_gk, top_n=10)
top_features_mid = select_top_features(feature_importance_mid, top_n=10)

# Tüm önemli özellikleri birleştirme
selected_features = set(top_features_def) | set(top_features_fwd) | set(top_features_gk) | set(top_features_mid)
selected_features = list(selected_features)

print("Selected Features:\n", selected_features)
```

```python
from sklearn.model_selection import train_test_split, cross_val_predict 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.svm import SVR, LinearSVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import classification_report, accuracy_score, recall_score, precision_score, confusion_matrix, roc_curve, roc_auc_score
from sklearn.metrics import confusion_matrix  
import warnings
warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')

X_selected = player_data[selected_features + ['Position_encoded']]
y = player_data['Value']

# K-Fold Cross-Validation fonksiyonu
def k_fold_cross_validation(X, y, model, num_folds=10):
    num_instances = len(X)
    fold_size = num_instances // num_folds

    indices = np.arange(num_instances)
    np.random.seed(123)
    np.random.shuffle(indices)

    X_shuffled = X.iloc[indices].reset_index(drop=True)
    y_shuffled = y.iloc[indices].reset_index(drop=True)

    best_model = None  
    best_score = float('-inf')
    scores = []
    predictions = []
    mse_scores = []
    mae_scores = []
    r2_scores = []

    for i in range(num_folds):
        start = i * fold_size
        end = (i + 1) * fold_size if i < num_folds - 1 else num_instances

        X_test_fold = X_shuffled.iloc[start:end]
        y_test_fold = y_shuffled.iloc[start:end]
        X_train_fold = pd.concat([X_shuffled.iloc[:start], X_shuffled.iloc[end:]], axis=0)
        y_train_fold = pd.concat([y_shuffled.iloc[:start], y_shuffled.iloc[end:]], axis=0)

        model.fit(X_train_fold, y_train_fold)
        y_pred = model.predict(X_test_fold)

        score = model.score(X_test_fold, y_test_fold)
        scores.append(score)
        predictions.extend(y_pred)

        mse = mean_squared_error(y_test_fold, y_pred)
        mae = mean_absolute_error(y_test_fold, y_pred)
        r2 = r2_score(y_test_fold, y_pred)

        mse_scores.append(mse)
        mae_scores.append(mae)
        r2_scores.append(r2)

        if score > best_score:
            best_score = score
            best_model = model

    print("Best Model Score:", best_score)
    print("Average MSE Score:", np.mean(mse_scores))
    print("Average MAE Score:", np.mean(mae_scores))
    print("Average R2 Score:", np.mean(r2_scores))

    return best_model, scores, predictions, mse_scores, mae_scores, r2_scores

# Model sözlüğü oluşturma
models = {
    'RandomForestRegressor': RandomForestRegressor(),
    'KNeighborsRegressor': KNeighborsRegressor(),
    'AdaBoostRegressor': AdaBoostRegressor(),
    'SVR': SVR(),
    'LinearSVR': LinearSVR(),
    'MLPRegressor': MLPRegressor(),

}

# Her bir model için cross-validation sonuçlarını yazdırma
for model_name, model in models.items():
    print(f"{model_name} modeli için cross-validation sonuçları:")
    best_model, scores, predictions, mse_scores, mae_scores, r2_scores = k_fold_cross_validation(X_selected, y, model)
    print("\n---------------------------------------------------------------------\n")
```
